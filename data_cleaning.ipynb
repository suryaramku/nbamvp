```python
players = pd.read_csv("players.csv")
players.columns = map(str.lower, players.columns)
players
```

```python
# players['player'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
```

```python
del players["unnamed: 0"]
del players["rk"]
```

```python
players.head()
```

```python
# players["player"].head(50)
```

```python
players["player"] = players["player"].str.replace("*", "", regex = False)
```

```python
# want each row to have one player, take greg anderson as an example
```

```python
players.groupby(["player", "year"]).get_group(("Greg Anderson" , 1991))
```

```python
def single_row(df):
    if df.shape[0] ==1: # only one row in the group, dont do anything
        return df
    else: # otherwise we need to change
        row = df[df["tm"] == "TOT"] # we dont care about other rows, we just want row with TOT as their team
        row["tm"] = df.iloc[-1,:]["tm"] # replace the tm with the last team he played for , select last row, and select the Tm
        return row 
players = players.groupby(["player", "year"]).apply(single_row)
```

```python
players.head(20)
```

```python
players.index = players.index.droplevel()
```

```python
players
```

```python
combined = players.merge(mvps, how = "outer", on = ["player", "year"])  # need to do an outer merge, need to keep all data
combined
```

```python
combined[combined["pts won"] > 0]
```

```python
combined[["pts won", "pts max", "share"]] = combined[["pts won", "pts max", "share"]].fillna(0)
```

```python
combined
```

## teams dataframe

```python
teams = pd.read_csv("standings.csv")
teams.columns = map(str.lower,teams.columns)
teams.head(30)
```

```python
teams = teams[~teams["w"].str.contains("Division")] # removes rows that have a string with division
```

```python
teams["team"] = teams["team"].str.replace("*", "", regex = False) # remove asterick
```

```python
nicknames = {}

with open("nick.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev, name = line.replace("\n", "").split(",")
        nicknames[abbrev] = name
        
        
```

```python
nicknames
```

```python
combined["team"] = combined["tm"].map(nicknames)
```

```python
combined.head()
```

```python
stats = combined.merge(teams, how = "outer", on = ["team", "year"])
stats
```

```python
del stats["unnamed: 0"]
```

```python
# stats.dtypes # always good to check this
stats = stats.apply(pd.to_numeric, errors = "ignore")
```
