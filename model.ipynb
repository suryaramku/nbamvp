```python
predictors = [ 'age', 'g', 'gs', 'mp', 'fg', 'fga', 'fg%', '3p',
       '3pa', '3p%', '2p', '2pa', '2p%', 'efg%', 'ft%', 'orb',
       'drb', 'trb', 'ast', 'stl', 'blk', 'tov', 'pf', 'pts',
        'w/l%', 'gb', 'ps/g',
       'pa/g', 'srs'] 
```

```python
train = stats[stats["year"] < 2022]
```

```python
test = stats[stats["year"] == 2022]
```

```python
from sklearn.linear_model import Ridge
# prevents overfitting and shrinks linear regression coefficients
reg = Ridge(alpha = .1) # how much coefficient is shrunk
reg.fit(train[predictors], train ["share"])
```

```python
predictions = reg.predict(test[predictors])
```

```python
predictions = pd.DataFrame(predictions, columns = ["predictions"], index = test.index)
predictions
```

```python
combinations = pd.concat([test[["player", "share"]], predictions], axis = 1)
```

```python
combinations
```

```python
combinations.sort_values("share", ascending = False).head(10)
```

```python
# error metric

from sklearn.metrics import mean_squared_error

mean_squared_error(combinations["share"], combinations["predictions"]) # mean difference
```

```python
combinations = combinations.sort_values("share", ascending = False)
combinations["rk"] = list(range(1, combinations.shape[0] + 1))

combinations
```

```python
combinations = combinations.sort_values("predictions", ascending = False)
combinations["predicted rk"] = list(range(1, combinations.shape[0] + 1))

combinations
```

```python
combinations.sort_values("share", ascending = False).head(10)
combinations[combinations["predicted rk"] <= 5]
```

```python
def find_ap(combinations):
    actual = combinations.sort_values("share", ascending=False).head(5)
    predicted = combinations.sort_values("predictions", ascending=False)
    ps = []
    found = 0
    seen = 1
    for index,row in predicted.iterrows():
        if row["player"] in actual["player"].values:
            found += 1
            ps.append(found / seen)
        seen += 1

    return  sum(ps) / len(ps)
```

```python
find_ap(combinations) # this is your error metric
```
