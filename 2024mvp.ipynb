```python
import pandas as pd
```

```python
data24 = pd.read_csv("2024data.csv")
```

```python
data24.rename(columns = {"winloss": 'w/l%' }, inplace = True)
a = data24[data24["player"] == "Player"]
data24 = data24.drop(a.index)
data24
```

```python
data24 = data24.apply(pd.to_numeric, errors = "ignore")
```

```python
stats = pd.read_csv("allstats.csv")
stats
```

```python
del stats["Unnamed: 0"]
```

```python
data24 = data24.fillna(0)
data24
```

```python
data24.columns
```

```python
predictors  = [ 'g', 'gs', 'mp', 'fg', 'fg%', '3p%', '2p%', 'efg%', 'ft%', 'orb', 'drb', 'trb', 'ast', 'stl', 'blk', 'tov', 'pf',
       'pts', 'w/l%']
```

```python
train = stats[stats["year"] < 2023]
```

```python
test = data24
```

```python
stats = stats.fillna(0)
stats
```

```python
from sklearn.linear_model import Ridge
# prevents overfitting and shrinks linear regression coefficients
reg = Ridge(alpha = .1) # how much coefficient is shrunk
reg.fit(train[predictors], train ["share"])
```

```python
predictions = reg.predict(test[predictors])
```

```python
predictions = pd.DataFrame(predictions, columns = ["predictions"], index = test.index)
predictions
```

```python
combinations = pd.concat([test[["player"]], predictions], axis = 1)
combinations
```

```python
top5 = combinations.sort_values("predictions", ascending = False).head(10)
top5
```

```python
import numpy as np
a = top5.take(np.arange(5))
a = a.reset_index()
a
```
